```python
import os
import pickle

# Define the directory containing the pickle files
directory_path = path_to_main_folder + 'data/'

# Iterate over files in the directory
for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)

    # Check if the file is a pickle file (ends with .pkl)
    if filename.endswith('.pickle'):
        try:
            # Open and read the pickle file
            with open(file_path, 'rb') as file:
                data = pickle.load(file)



            data['actual_requests']= 0

            # Setting the counter to 0 & saving

            with open(directory_path+filename, 'wb') as file:
                pickle.dump(data, file)

            print(f"Processing {filename}: {data}")


        except Exception as e:
            print(f"Error processing {filename}: {str(e)}")
```
