```python
import json
import os
import csv

directory_path = "/home/hinux/Desktop/google_images/tester_images/true_images"
directory_path_best = "/home/hinux/Desktop/google_images/tester_images/best_test/Ve" # change path
oursList = os.listdir(directory_path)
best_list = os.listdir(directory_path_best)
finalList = list(set(best_list) & set(oursList))

```

```python
finalList
```

```python

data_dict = {"image_name": [],
             "score_our": [],
             "score_sheng": []
             }


for filename in finalList:
    if filename.endswith(".json"):
        with open(directory_path + "/" + filename) as json_file:
            data = json.load(json_file)
        temp = data["save_path"].split("/")[-1]           
        data_dict["image_name"].append(temp[:-4])
        data_dict["score_our"].append(data["score"])
        
        
# for best checkpoint


for filename in finalList:
    if filename.endswith(".json"):
        with open(directory_path_best + "/" + filename) as json_file:
            data = json.load(json_file)
        temp = data["save_path"].split("/")[-1][:-4]
        if temp in data_dict["image_name"]: 
            # data_dict["image_name"].append(temp[:-4])
            data_dict["score_sheng"].append(data["score"])
```
