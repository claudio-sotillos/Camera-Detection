```python
plt.figure(figsize=(19, 6))
x=data['epoch']
y=data['mAP']
plt.scatter(x,y)
plt.xlabel('Epochs',fontsize=25)
plt.ylabel('Mean Average Precision', fontsize=25)
plt.xticks(np.arange(0, 39, 1),fontsize=15)
plt.yticks(fontsize=15)
plt.title("mAP progress along epochs",fontsize=25 )
plt.plot(x, y, color='red', linestyle='-', linewidth=2, label='Line Plot')
plt.show()
```

```python

plt.figure(figsize=(15, 6))
x = data['epoch']
y = data['val_loss']
plt.scatter(x, y)
plt.title('Val_loss', fontsize=20)
plt.xlabel('epochs',fontsize=15)
plt.ylabel('val_loss', fontsize=15)
plt.xticks(np.arange(0, 39, 1))
plt.plot(x, y, color='red', linestyle='-', linewidth=2, label='Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming x and data are defined somewhere in your code
# x = ...
# data = ...

fig, ax1 = plt.subplots(figsize=(18, 6))
color = 'tab:red'
ax1.set_xlabel('Epochs',fontsize=25)
ax1.set_ylabel('Train_loss', color=color, fontsize=25)
ax1.plot(x, data['train_loss'], color=color)
ax1.tick_params(axis='y', labelcolor=color, labelsize=18)  # Set fontsize for y-axis ticks

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Val_loss', color=color, fontsize=25)  # we already handled the x-label with ax1
ax2.plot(x, data['val_loss'], color=color)
ax2.tick_params(axis='y', labelcolor=color, labelsize=18)  # Set fontsize for y-axis ticks

ax1.tick_params(axis='x', labelsize=15)  # Set fontsize for x-axis ticks
# ax2.tick_params(axis='x', labelsize=10)  # Set fontsize for x-axis ticks
plt.xticks(np.arange(0, 39, 1))
plt.title("Loss", fontsize=25)

plt.legend()
plt.show()
```
