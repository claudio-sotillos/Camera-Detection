# Getting the images from path

```python
import shutil

# List of image file paths
image_paths_list = list(grouped_strings.items()) # PASSING THE LIST HERE
T = "camera_train"
V = "camera_valid"
# Destination folder where you want to copy the images
destination_folder = f'/home/hinux/Desktop/project04/checking_folder/filtered_img_camera/{T}' # REMEMBER TO CHANGE THIS #####################################
# destination_folder = "/home/hinux/Desktop/project04/checking_folder/filtered_img_camera/train_val_images" # REMEMBER TO CHANGE THIS #####################################



counter = 0
for i in range(100,591): # 0,353 # 353,441  #REMEMBER TO CHANGE THE SIZE OF THE DATA SET HERE #####################################################3
    try:
        # Extract the file name from the image path
        file_name_0 = os.path.basename(image_paths_list[i][1][0])
        file_name_1 = os.path.basename(image_paths_list[i][1][1])

        # Construct the destination path by joining the destination folder and file name
        destination_path_0 = os.path.join(destination_folder, file_name_0)
        destination_path_1 = os.path.join(destination_folder, file_name_1)

        # print(destination_path_0)    
        # print(destination_path_1)    
        counter += 1

        # Copy the image to the destination folder
        shutil.copy(image_paths_list[i][1][0], destination_path_0)
        shutil.copy(image_paths_list[i][1][1], destination_path_1)

        print(f"Image {file_name_0} copied to {destination_folder}", counter)
        print(f"Image {file_name_1} copied to {destination_folder}", counter)
    except Exception as e:
        print(f"Error copying image: {str(e)}")

print(counter)
```
